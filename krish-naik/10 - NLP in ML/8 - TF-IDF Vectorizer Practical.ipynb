```python
import pandas as pd 
message = pd.read_csv("spam.csv",encoding='latin-1',sep=',')
message = message.iloc[:,[0,1]]
message.columns = ["label","message"]
```

```python
message.head()
```

```python
import re 
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
wordLem = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(message)):
    result = re.sub('[^a-zA-Z]',' ',message['message'][i])
    result = result.lower()
    result = result.split()
    result = [wordLem.lemmatize(word) for word in result if word not in set(stopwords.words('english'))]
    result = ' '.join(result)
    corpus.append(result)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=100)
x = tfidf.fit_transform(corpus).toarray()
```
