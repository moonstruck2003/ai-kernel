```python
#PorterStemmer
import nltk
from nltk.stem import PorterStemmer
```

```python
words = ["eating","eats","writting","writes","programming","programs","history"]
```

```python
porter = PorterStemmer()
```

```python
for word in words:
    print(word + "---> "+porter.stem(word))
```

```python
###RegexStemmer
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|es$|able$|',min=4)
```

```python
reg_stemmer.stem("carsers")
```

```python
#snowball stmemer --> better in compared to PorterStemmer 
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowball=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowball.stem(word))
```

```python
porter.stem("Fairly"),porter.stem("Sportingly")
```

```python
snowball.stem("Fairly"),snowball.stem("Sportingly") #Better output
```
