```python
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import SnowballStemmer
sbs = SnowballStemmer("english")
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
paragraph = """The brave explorers were continuously hiking through the dense, unforgiving forests. They quickly realized that their heavy backpacks were causing immense exhaustion. However, the stunning mountains in the distance kept them moving forward. Survival depends on making smart decisions and constantly adapting to changing environments!"""
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## apply StopWords  and Filter and then apply Stemming. 
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) #converting all the words into sentences 
```

```python
sentences
```

```python
## apply StopWords  and Filter and then apply Stemming.  Using SnowballStemmer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[sbs.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) #converting all the words into sentences 
```

```python
sentences
```

```python
## apply StopWords  and Filter and then apply Lemmatization.  Using WordNetLemmatizer
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemma.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) #converting all the words into sentences 
```
