```python
import nltk
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
```

```python
wordnet = WordNetLemmatizer()
```

```python
wordnet.lemmatize("goes")
```

```python
words = ["eating","eats","writting","writes","programming","programs","history"]
```

```python
for word in words:
    print(word + "-----> " + wordnet.lemmatize(word))
```

```python
wordnet.lemmatize("Fairly",pos='v'), wordnet.lemmatize("Sportingly",pos='v') #not changing the meaning
```
