```python
#EXAMPLE

from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.tokenize import TreebankWordTokenizer
```

```python
corpus = """
    Hello this is Mridul. Currently learning AI Stack course, by krish naik. Interested in AI/ML!! Hoping to create a career.
"""
```

```python
document = sent_tokenize(corpus)
for sentence in document:
    print(word_tokenize(sentence))
```

```python
treebanktokenizer = TreebankWordTokenizer()
```

```python
for vocabulary in document:
    print(treebanktokenizer.tokenize(vocabulary))
```
