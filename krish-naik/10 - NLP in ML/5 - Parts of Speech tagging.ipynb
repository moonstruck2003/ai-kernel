```python
paragraph = """The brave explorers were continuously hiking through the dense, unforgiving forests. They quickly realized that their heavy backpacks were causing immense exhaustion. However, the stunning mountains in the distance kept them moving forward. Survival depends on making smart decisions and constantly adapting to changing environments!"""
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
sentence = nltk.sent_tokenize(paragraph)
```

```python
from nltk.corpus import stopwords
```

```python
##we will find the pos tag 

for i in range(len(sentence)):
    words = nltk.word_tokenize(sentence[i])
    word = [word for word in words if word not in set(stopwords.words('english'))]
    pos_tag = nltk.pos_tag(word)
    print(pos_tag)     
```
